```python
# Loop through the TLE data to create a SatrecArray with several individual space objects 

satellites = []  # SatrecArray to store all space objects 

for i in range(0, len(TLE_sample), 2):
    line1 = TLE_sample[i]
    line2 = TLE_sample[i+1]
    
    # Check to see if the Elset Classification is 'U' indicating that it is an unclassified object, that is, a space debris 
    if TLE_sample[i].split()[1][-1] == 'U':
        satellites.append(Satrec.twoline2rv(line1, line2))

satA = SatrecArray(satellites)
```

```python
# Handle the Julian dates for an arbitrary date

jd, fr = jday(2020, 8, 20, 12, 0, 0)

# Convert to numpy array 
jd = np.array([jd])
fr = np.array([fr])
```

```python
# Use sgp4 to compute the error, position (TEME: True Equator Mean Equinox), and velocity

err, r, v = satA.sgp4(jd, fr)
```

```python
# Export the data to a JSON file 

res_dict = {'Error': err.tolist(), 'Postion in TEME (km)': r.tolist(), 'Velocity in TEME (km/s)': v.tolist()}

file_name = input('Name your JSON file -> ')
if os.path.exists(file_name):
    os.remove(file_name)
with open(file_name, 'w') as jfile:
    json.dump(res_dict, jfile, indent=4)
```

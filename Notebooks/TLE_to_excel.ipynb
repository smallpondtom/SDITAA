```python
# Organize the data so that we can convert the data to a numpy array and then to a dataframe 

text_by_line = TLE_sample
text_by_element = []
for i in range(0, len(text_by_line), 2):
    # Split each lines by delimiter of spaces 
    temp1 = text_by_line[i].split()    
    temp2 = text_by_line[i+1].split()
    del temp1[0]
    del temp2[0]
    
    # Manipulate the first element to separate into satellite catalog number and elset classification number 
    this = re.split('(\d+)', temp1[0])
    del this[this.index('')]
    del temp1[0]
    temp1.insert(0, this[0])
    temp1.insert(1, this[1])
    
    # Remove the redundant satellite catalog number in the second line
    del temp2[0]
    
    # Remove the checksum 
    temp1[-1] = temp1[-1][:-1]
    temp2[-1] = temp2[-1][:-1]
    
    # Separate the last element in temp2 into Mean Motion and Revolution Number at Epoch
    mm = [temp2[-1][:-5]]
    rne = [temp2[-1][-5:]]
    del temp2[-1]
    temp2 += mm
    temp2 += rne
    
    # If there is no international Designator insert a string of 'NaN'
    if len(temp1) != 9:
        temp1.insert(2, 'NaN')
    
    # Delete any NoneType in the list
    if '' in temp1:
        del temp1[temp1.index('')]
    if '' in temp2:
        del temp2[temp2.index('')]
    
    text_by_element.append(temp1 + temp2)
```

```python
# Convert array to numpy array 

data = np.asarray(text_by_element, dtype=object)
data.shape  # Check the shape
```

```python
# Create a pandas dataframe from the data 

df = pd.DataFrame(data=data, columns=['Satellite Catalog Number', 'Elset Classification', 'International Designator', 'Element Set Epoch (UTC)',
                                      '1st Derivative of the Mean Motion with respect to Time', '2nd Derivative of the Mean Motion with respect to Time (decimal point assumed)',
                                      'B* Drag Term', 'Element Set Type', 'Element Number', 'Orbit Inclination (deg)', 'Right Ascension of Ascending Node (deg)',
                                      'Eccentricity (decimal point assumed)', 'Argument of Perigee (deg)', 'Mean Anomally (deg)', 'Mean Motion (rev/day)', 'Revolution Number at Epoch'])
```

```python
# Change the B* Drag values into float numbers 

for i, val in enumerate(df['B* Drag Term']):
    num = float(val[:-2])
    sup = int(val[-1])
    
    if val[-2] == '-':
        bdrag = num * 10 ** (-sup)
    else:
        bdrag = num * 10 ** (sup)
    df['B* Drag Term'].iloc[i] = bdrag 
    
# Change the 2nd Derivative of Mean motion into float numbers 

for i, val in enumerate(df['2nd Derivative of the Mean Motion with respect to Time (decimal point assumed)']):
    num = float(val[:-2])
    sup = int(val[-1])
    
    if val[-2] == '-':
        temp = num * 10 ** (-sup)
    else:
        temp = num * 10 ** (sup)
    df['2nd Derivative of the Mean Motion with respect to Time (decimal point assumed)'].iloc[i] = temp 
```

```python
# Change the Element Set Epoch to year, day, hour, min, second

def UTC2datetime(epoch):
    # Find the year
    two_digit = int(epoch[:2])
    if two_digit < 57:
        year = 2000 + two_digit
    else:
        year = 1900 + two_digit
        
    day_of_year = float(epoch[2:])
    # Find the day
    day = int(day_of_year) 
    # Find the hour 
    dec1 = day_of_year - day
    hour_of_year = dec1 * 24.0
    hour = int(hour_of_year)
    # Find the minute
    dec2 = hour_of_year - hour
    minute_of_year = dec2 * 60
    minute = int(minute_of_year)
    # Find the second
    dec3 = minute_of_year - minute
    second = dec3 * 60
    return year, day, hour, minute, second

datetime = np.zeros((df.shape[0], 5))
for i, x in enumerate(df['Element Set Epoch (UTC)']):
    datetime[i, :] = UTC2datetime(x)
```

```python
# Add a column with the datetimes

datetime_label = ['Year', 'Day', 'Hour', 'Minute', 'Second']
for i, label in enumerate(datetime_label):
    df[label] = datetime[:, i]
```
